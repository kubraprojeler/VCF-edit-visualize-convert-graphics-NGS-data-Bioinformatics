```python
#somatic ve strelka tablolarının içindeki DP(okuma derinliği) değerlerine ihtiyacım var. birçok değer tabloda aynı 
#sütunda toplu halde satıra yazılmış olduğu için ve ":" işareti ile ayrıldığı için istediğim değeri aşağıdaki kodla seçtim.
```

```python
for q in list_somatic:
        z=q[["NORMAL","TUMOR"]]
        z["DPNORMAL"]=z["NORMAL"].str.split(":",expand=True)[2]
        z["DPTUMOR"]=z["TUMOR"].str.split(":",expand=True)[2]
        a=z[["DPTUMOR"]]
        b=z[["DPNORMAL"]]
        q["DPNORMAL"]=b["DPNORMAL"]
        q["DPTUMOR"]=a["DPTUMOR"]
        q["DPNORMAL"]=pd.to_numeric(q["DPNORMAL"])
        q["DPTUMOR"]=pd.to_numeric(q["DPTUMOR"])
```

```python
for q in list_strelka:
        z=q[["NORMAL","TUMOR"]]
        z["DPNORMAL"]=z["NORMAL"].str.split(":",expand=True)[0]
        z["DPTUMOR"]=z["TUMOR"].str.split(":",expand=True)[0]
        a=z[["DPTUMOR"]]
        b=z[["DPNORMAL"]]
        q["DPNORMAL"]=b["DPNORMAL"]
        q["DPTUMOR"]=a["DPTUMOR"]
        q["DPNORMAL"]=pd.to_numeric(q["DPNORMAL"])
        q["DPTUMOR"]=pd.to_numeric(q["DPTUMOR"])
```
